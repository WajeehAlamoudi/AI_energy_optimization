```python
import requests
import pandas as pd

# Example: Istanbul coordinates
url = "https://archive-api.open-meteo.com/v1/archive"
params = {
    "latitude": 41.0082,
    "longitude": 28.9784,
    "start_date": "2024-01-01",
    "end_date": "2024-12-31",
    "hourly": "temperature_2m,relative_humidity_2m,wind_speed_10m,precipitation"
}
r = requests.get(url, params=params)
data = r.json()

weather_df = pd.DataFrame(data["hourly"])
weather_df.head()
```

```python
weather_df.to_csv("outdoor_weather.csv", index=False)
```

```python
!kaggle datasets download -d uciml/electric-power-consumption-data-set
!unzip electric-power-consumption-data-set.zip -d energy_data

energy_df = pd.read_csv("energy_data/household_power_consumption.txt", sep=';',
                        parse_dates={'datetime': ['Date','Time']},
                        infer_datetime_format=True, low_memory=False)
energy_df.head()
```

```python
import pandas as pd
import numpy as np

# Replace '?' with NaN and convert to numeric
for col in ['Global_active_power', 'Global_reactive_power', 'Voltage',
            'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']:
    energy_df[col] = pd.to_numeric(energy_df[col], errors='coerce')

# Drop any rows that have NaN in these columns
energy_df = energy_df.dropna(subset=['Global_active_power'])
```

```python
# Note: use '1h' instead of '1H' to remove the warning
energy_df = energy_df.set_index("datetime").resample("1h").mean().reset_index()
```

```python
energy_df.head()
energy_df.info()
```

```python
weather_df['time'] = pd.to_datetime(weather_df['time'])
energy_df['datetime'] = pd.to_datetime(energy_df['datetime'])
```

```python
# Resample with lowercase 'h'
energy_df = energy_df.set_index("datetime").resample("1h").mean().reset_index()

# Now merge cleanly
merged = energy_df.merge(weather_df, left_on="datetime", right_on="time", how="inner")
merged.head()
```

```python
merged.to_csv("combined_energy_weather_data.csv", index=False)
```
